Log Transformation

```python
def log_transfom_clean(x):
    x = np.log10(x)
    if np.isfinite(x):
        return x
    else:
        return 0

CLV_log = data['Customer Lifetime Value'].apply(lambda x :log_transfom_clean(x))
sns.displot(CLV_log, kde=True)
plt.show()

Income_log = data['Income'].apply(lambda x :log_transfom_clean(x))
sns.displot(Income_log, kde=True)
plt.show()

MPA_log = data['Monthly Premium Auto'].apply(lambda x :log_transfom_clean(x))
sns.displot(MPA_log, kde=True)
plt.show()
```

Box Cox Transformation

```python
#  filter out negative values:
len(data[data['Income']<=0])  # number of such observations.
# # if you do not want to lose this data, replace it with the mean of the column
data['Income_Transform'] = np.where(data['Income']<=0,0,data['Income'])

# # now calculate mean of the new column , excluding zeros in the count
mean = np.sum(data['Income_Transform'])/len(data[data['Income_Transform']>0])

data['Income_Transform'] = data['Income_Transform'].replace(0,mean)
data = data.drop(['Income'], axis=1)
sns.displot(data['Income_Transform'])
plt.show()

#Not good results because mean value is overreprresented.
```

```python
from scipy import stats
xt, lmbda = stats.boxcox(data['Income_Transform'])
sns.displot(xt)
plt.show()

```

```python
xt, lmbda = stats.boxcox(data['Customer Lifetime Value'])
sns.displot(xt)
plt.show()

# Box cox transformation improves the below variables to more normal distributions.
```

```python
xt, lmbda = stats.boxcox(data['Monthly Premium Auto'])
sns.displot(xt)
plt.show()
```
